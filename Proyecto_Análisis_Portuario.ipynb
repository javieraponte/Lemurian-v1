# Proyecto de análisis del trafico portuario en Colombia

## Introducción

---

### Contexto

La logistica portuaria es el conjunto de actividades y procesos necesarios para asegurar el flujo eficiente de bienes, mercancias y servicios a través de los puertos maritimos. Desde la gestión de la carga y descarga hasta la distribución y transporte, esta disciplina se encarga de optimizar cada etapa en la cadena logística.
Colombia, con su privilegiada posición geográfica con acceso tanto al Océano Pacífico como al Mar Caribe, se posiciona como un actor relevante en este ámbito.

### Objetivo

Para una empresa logistica es fundamental realizar seguimiento constante de todos los articulos en su inventario, en este caso puntual, una empresa logistica en los puertos es escencial esta tarea con cada cargamento que se mueve en el puerto maritimo.

El objetivo de este proyecto es evidenciar aspectos importantes que ayuden a direccionar a la empresa a tomar decisiones criticas relacionadas con la planificación, rentabilidad, eficiencia e inversión.

### Preguntas problema

Las siguientes preguntas ayudaran a direccionar las consultas para un analisis optimo segun el objetivo planteado

1. ¿Cual es el puerto maritimo con mayor trafico maritimo en los ultimos cinco años?
2. ¿Cual es el tipo de carga con mas recurrencia en cada puerto y cuanta es su carga anual?
3. ¿Cuanta carga se moviliza en cabotaje anual y que relación tiene con el trafico de carga expo/impo?
4. ¿Que porcentaje de carga es transitoria con respecto al trafico de carga expo/impo?

## Entendimiento de los datos

---

### Obtención de los datos

El conjunto de datos utilizado para este proyecto es [Trafico Portuario Marítimo En Colombia](https://www.datos.gov.co/d/5r3g-zv5z) y contiene información sobre la ciudad donde se ubica el puerto, la empresa prestadora del servicio en el puerto, el tipo de carga, las toneladas de cada carga, transbordos, cabotaje y año.

### Descripción de los datos

A continuación se realizara una exploración y descripción inicial de cada conjunto de datos para mostrar cada una de sus atributos y tipo de valores de como vienen incialmente.

* **Zona portuaria:**
  
  Lugar donde se encuentra ubicado el puerto.
  - Tipo TEXTO

* **Sociedad portuaria:**

  Empresa prestadora de los servicios dentro de los puertos.
  - Tipo TEXTO

* **Tipo de servicio:**

  Si el servicio pertenece a una entidad publica o privada.
  - Tipo TEXTO

* **Tipo de carga:**

  Nombre clasificada de la carga.
  - Tipo TEXTO

* **Exportación:**

  Trafico de carga de exportación definida en toneladas.
  - Tipo NÚMERO

* **Importación:**

  Trafico de carga de importación definida en toneladas.
  - Tipo NÚMERO

* **Transbordo:**

  La carga cambia de medio de transporte.
  - Tipo NÚMERO

* **Transito internacional:**

  Trafico de otros paises.
  - Tipo NÚMERO

* **Fluvial:**

  Transito de personas, animales o cosas definida en toneladas por medio de embarcaciones que navegan por rios, canales, lagos, lagunas, cienagas, etc.
  - Tipo NÚMERO

* **Cabotaje:**

  Transito de carga definida en toneladas que se movilizan entre puertos sin apartarse de la costa.
  - Tipo NÚMERO

* **Movilizaciónes a bordo:**

  Cantidad de movilizaciones de carga dentro de un buque.
  - Tipo NÚMERO

* **Transitoria:**

  Cantidad de carga ubicada en una zona temporal dentro del buque antes de ser redistribuida o descargada.
  - Tipo NÚMERO

* **Año vigencia:**

  Año en que se genero la información.
  - Tipo NÚMERO

* **Mes vigencia:**

  Mes en que se genero la información.
  - Tipo NÚMERO
